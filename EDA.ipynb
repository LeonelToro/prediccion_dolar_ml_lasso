```python
import pandas as pd

# Cargar el DataFrame mensual procesado
path = r"C:\Users\primo\Desktop\Ciencia de Datos\Proyectos\BCRA\data\processed\consolidado.csv"
df = pd.read_csv(path, parse_dates=["fecha"])

# Mostrar las primeras filas
df.head()
```

```python
df.info()
df.describe().T
```

```python
# Cantidad de valores nulos por columna
nulos = df.isna().sum()

# Porcentaje de nulos
porcentaje_nulos = (nulos / len(df)) * 100

# Unir en un DataFrame para verlo ordenado
df_nulos = pd.DataFrame({
    "nulos": nulos,
    "porcentaje": porcentaje_nulos
}).sort_values(by="porcentaje", ascending=False)

df_nulos
```

```python
# Matriz de correlación Pearson entre variables numéricas
corr = df.corr()

# Mostrar matriz completa
corr
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ajustamos tamaño del gráfico
plt.figure(figsize=(12, 10))

# Graficar heatmap con anotaciones y paleta de colores
sns.heatmap(corr, annot=True, fmt=".2f", cmap="cividis", center=0, square=True, linewidths=0.5)

# Título y ajuste para que no se sobreponga nada
plt.title("Mapa de calor de correlaciones entre variables")
plt.tight_layout()

plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))

# Eje izquierdo: inflación
color1 = "tab:red"
ax1.set_xlabel("Fecha")
ax1.set_ylabel("Inflación mensual (%)", color=color1)
ax1.plot(df["fecha"], df["inflacion"], color=color1, marker="o", label="Inflación")
ax1.tick_params(axis="y", labelcolor=color1)

# Eje derecho: dólar oficial
ax2 = ax1.twinx()
color2 = "tab:blue"
ax2.set_ylabel("USD oficial ($ARS)", color=color2)
ax2.plot(df["fecha"], df["usd_oficial"], color=color2, marker="o", label="USD oficial")
ax2.tick_params(axis="y", labelcolor=color2)

plt.title("Inflación mensual vs. dólar oficial")
fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))

color1 = "tab:purple"
ax1.set_xlabel("Fecha")
ax1.set_ylabel("Brecha USD (%)", color=color1)
ax1.plot(df["fecha"], df["brecha_usd"], color=color1, label="Brecha USD")
ax1.tick_params(axis="y", labelcolor=color1)

ax2 = ax1.twinx()
color2 = "tab:green"
ax2.set_ylabel("LELIQ", color=color2)
ax2.plot(df["fecha"], df["leliq"], color=color2, label="LELIQ")
ax2.tick_params(axis="y", labelcolor=color2)

plt.title("Brecha cambiaria vs. LELIQ")
fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(12, 5))

color1 = "tab:brown"
ax1.set_xlabel("Fecha")
ax1.set_ylabel("Reservas ($M USD)", color=color1)
ax1.plot(df["fecha"], df["reservas"], color=color1, label="Reservas")
ax1.tick_params(axis="y", labelcolor=color1)

ax2 = ax1.twinx()
color2 = "tab:blue"
ax2.set_ylabel("USD general", color=color2)
ax2.plot(df["fecha"], df["usd_general"], color=color2, label="USD general")
ax2.tick_params(axis="y", labelcolor=color2)

plt.title("Reservas internacionales vs. dólar paralelo (general)")
fig.tight_layout()
plt.show()
```

```python
df_plot = df[["fecha", "base_monetaria", "inflacion"]].dropna()

fig, ax1 = plt.subplots(figsize=(12, 5))

color1 = "tab:blue"
ax1.set_xlabel("Fecha")
ax1.set_ylabel("Base monetaria", color=color1)
ax1.plot(df_plot["fecha"], df_plot["base_monetaria"], color=color1)
ax1.tick_params(axis="y", labelcolor=color1)

ax2 = ax1.twinx()
color2 = "tab:red"
ax2.set_ylabel("Inflación (%)", color=color2)
ax2.plot(df_plot["fecha"], df_plot["inflacion"], color=color2)
ax2.tick_params(axis="y", labelcolor=color2)

plt.title("Base monetaria vs. inflación mensual")
fig.tight_layout()
plt.show()
```
